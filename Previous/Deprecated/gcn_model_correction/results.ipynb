```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_wt = pd.read_csv("results_243_5_wt.csv", index_col=0)
results_wot = pd.read_csv("results_243_5_wot.csv", index_col=0)
```

```python
epoch_start = 0

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
count = 0
for result in results_wt.columns:
    row = count//2
    col = count%2
    count += 1

    ax[row][col].plot(np.arange(epoch_start, len(results_wot[result])), results_wot[result][epoch_start:])
    ax[row][col].plot(np.arange(epoch_start, len(results_wt[result])), results_wt[result][epoch_start:])
    ax[row][col].set_xlabel(result)
    ax[row][col].legend(["No transitions", "Transitions"])
```
