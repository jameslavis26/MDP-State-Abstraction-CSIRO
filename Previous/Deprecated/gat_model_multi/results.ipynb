```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gap_729 = pd.read_csv("multi_run_729_states_10_runs/test_gap.csv", index_col=0)
times_729 = pd.read_csv("multi_run_729_states_10_runs/times.csv", index_col=0)
error_729 = pd.read_csv("multi_run_729_states_10_runs/test_error.csv", index_col=0)

gap_243 = pd.read_csv("multi_run_243_states_10_runs/test_gap.csv", index_col=0)
times_243 = pd.read_csv("multi_run_243_states_10_runs/times.csv", index_col=0)
error_243 = pd.read_csv("multi_run_243_states_10_runs/test_error.csv", index_col=0)

gap_81 = pd.read_csv("multi_run_81_states_10_runs/test_gap.csv", index_col=0)
times_81 = pd.read_csv("multi_run_81_states_10_runs/times.csv", index_col=0)
error_81 = pd.read_csv("multi_run_81_states_10_runs/test_error.csv", index_col=0)
```

```python
gap_81_stack = gap_81.stack()
gap_81_stack.index = gap_81_stack.index.to_flat_index().map(lambda x: x[0])
sns.lineplot(gap_81_stack, errorbar='ci')

gap_243_stack = gap_243.stack()
gap_243_stack.index = gap_243_stack.index.to_flat_index().map(lambda x: x[0])
sns.lineplot(gap_243_stack, errorbar='ci')

gap_729_stack = gap_729.stack()
gap_729_stack.index = gap_729_stack.index.to_flat_index().map(lambda x: x[0])
sns.lineplot(gap_729_stack, errorbar='ci')

plt.legend(["81", "", "243", "", "739"])
plt.ylabel("Training Loss")
plt.xlabel("Epoch")
plt.hlines(y=0.019396144577694768, xmax=300, xmin=0, color='r')
```
