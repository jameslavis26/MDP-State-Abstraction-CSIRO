```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gat_lr = [0.0001, 0.0005, 0.001]
gcn_lr = [0.0001, 0.0005, 0.001]
gat_lr_node = [0.001, 0.0008, 0.0006, 0.0004, 0.00005, 0.00001]

models = ["GCN_NoLR_test", "GAT_NoLR_test", "GAT_NoLR_test_node"]
results = ["training_loss", "test_loss", "test_gap", "test_error", "test_recall"]
```

```python
n_plots = 5
n_cols = 2
n_rows = int((n_plots + n_plots%2)/2)

fig, ax = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 6*n_rows))

model = "GCN_NoLR_test"

for i in range(len(results)):
    col = i %2
    row = i //2 
    
    legend = []
    for lr in gcn_lr:
        dataset=f"Reserve_MDP_243_7_lr_{lr*10:.4f}".replace(".", "")

        df = pd.read_csv(f"Results/{model}/{dataset}/{results[i]}.csv", index_col=0)

        df_long = df.stack()
        df_long.index = df_long.index.to_flat_index().map(lambda x: x[0])
        sns.lineplot(df_long, errorbar='ci', ax=ax[row, col])
        ax[row, col].set_ylabel(results[i])
        ax[row, col].set_xlabel("Epoch")
        legend.append(f"{dataset}")
        legend.append("")
    ax[row, col].legend(legend)
```
